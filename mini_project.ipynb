```python
# Sort colors
def sort_colors(color_string):
    # Split by hyphen into a list
    colors = color_string.split("-")
    # Sort alphabetically (works for both upper and lower case)
    colors.sort()
    # Join back with hyphens
    return "-".join(colors)

# Test both samples
input1 = "green-red-yellow-black-white"
input2 = "PINK-BLUE-TAN-PURPLE"

print(f"Input:  {input1}")
print(f"Output: {sort_colors(input1)}")
print()
print(f"Input:  {input2}")
print(f"Output: {sort_colors(input2)}")
```

```python
#Playing with names
def ispalindrome(name):
    """Check if the name reads the same forwards and backwards."""
    clean = name.replace(" ", "").lower()
    if clean == clean[::-1]:
        print("Yes it is a palindrome.")
    else:
        print("No it is not a palindrome.")

def count_the_vowels(name):
    """Count how many vowels are in the name."""
    vowels = "aeiou"
    count = sum(1 for ch in name.lower() if ch in vowels)
    print(f"No of vowels: {count}")

def frequency_of_letters(name):
    """Count how many times each letter appears (ignoring spaces)."""
    freq = {}
    for ch in name.lower():
        if ch != " ":
            freq[ch] = freq.get(ch, 0) + 1
    result = ", ".join(f"{ch}-{count}" for ch, count in freq.items())
    print(f"Frequency of letters: {result}")
```

```python
#Happiness
import sys

# Read three command line arguments
string1 = sys.argv[1]   # liked numbers
string2 = sys.argv[2]   # disliked numbers
string3 = sys.argv[3]   # numbers to evaluate

# Convert each hyphen-separated string into a set of numbers
liked    = set(string1.split("-"))
disliked = set(string2.split("-"))
given    = string3.split("-")

# Calculate happiness
happiness = 0
for num in given:
    if num in liked:
        happiness += 1
    elif num in disliked:
        happiness -= 1

print(f"Final happiness: {happiness}")
```

```python
#Find the secret
from collections import Counter
import re

def find_secret_message(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()

    # 1. Determine Meeting Time
    num_lines = len(lines)
    if num_lines > 12:
        meeting_time = f"{num_lines - 12} PM"
    else:
        meeting_time = f"{num_lines} AM"

    # 2. Determine Meeting Place
    # Combine all lines and find all words (ignoring punctuation/case)
    text = " ".join(lines).lower()
    words = re.findall(r'\b\w+\b', text)

    if words:
        most_common_word, _ = Counter(words).most_common(1)[0]
        meeting_place = f"{most_common_word.capitalize()} Street"
    else:
        meeting_place = "Unknown Street"

    # Output Results
    print(f"Meeting time: {meeting_time}")
    print(f"Meeting place: {meeting_place}")

# Usage
# find_secret_message('sample.txt')
```

```python
#Purchase
import os

def process_purchase():
    file_name = input("Enter the file name: ")

    # Ensure the file name has the correct extension if not provided
    if not file_name.endswith('.txt'):
        file_name += '.txt'

    try:
        with open(file_name, 'r') as file:
            lines = [line.strip() for line in file if line.strip()]

            if not lines:
                print("The file is empty.")
                return

            items_purchased = []
            discount = 0.0

            # Process lines
            for line in lines:
                parts = line.split()
                # Check if it's the discount line (usually the last one)
                if parts[0].lower() == "discount":
                    discount = float(parts[1])
                else:
                    # Item name and price
                    price = float(parts[-1])
                    items_purchased.append(price)

            # Calculations
            no_of_items = len(items_purchased)
            free_items = items_purchased.count(0)
            total_amount = sum(items_purchased)
            final_amount = total_amount - discount

            # Output results
            print(f"No of items purchased: {no_of_items}")
            print(f"No of free items: {free_items}")
            print(f"Amount to pay: {total_amount}")
            print(f"Discount given: {discount}")
            print(f"Final amount paid: {final_amount}")

    except FileNotFoundError:
        print(f"Error: The file '{file_name}' was not found.")
    except ValueError:
        print("Error: The file contains invalid data formats (ensure prices are numbers).")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    process_purchase()
```
